```python
b0=wdata['CommentsNumber']
def sumval(df):
  r=0
  for e in df.values:
    if type(e)==str:
       if len(e)>3:
          ldd=e.split()
          e=ldd[0]+ldd[1]
    r+=int(e)
  return r

sCMW=sumval(b0)
sCMW
```

```python
def intlisr(df):
  lis=[]
  for e in df.values:
    if type(e)==str:
       if len(e)>3:
          ldd=e.split()
          e=ldd[0]+ldd[1]
    lis.append(int(e))
  return lis
```

# Posts with hashtags

```python
ndata=data.loc[(data['Hashtags']!='[]')]
```

```python
ndata.isnull().sum().sort_values(ascending=False)
```

```python
ndata.fillna(0,inplace=True)
```

```python
b=ndata['CommentsNumber']
def sumval(df):
  r=0
  for e in df.values:
    if type(e)==str:
       if len(e)>3:
          ldd=e.split()
          e=ldd[0]+ldd[1]
    r+=int(e)
  return r

sCM=sumval(b)
sCM
```

```python
c=ndata['RetweetsNumber']
c[2]=26400
c.values
sRN=sumval(c)
sRN
```

```python
d=ndata['LikesNumber']
print(d.values)
d[0]=16700
d[2]=265000
sLN=sumval(d)
sLN
```

```python
lp=["Posts with Hashtags","Posts without Hashtags"]
cn=[sCM/60,sCMW/15]
rn=[sRN/60,sRNW/15]
ln=[]
```
